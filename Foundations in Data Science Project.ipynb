### Analysis on different food prices and comparison of different food items to make informed predictions about the future.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ng=pd.read_csv("cfafrica-_-data-team-_-wanadata-_-nigeria-_-food-prices-_-april-2020-national.csv")

#reading top 5 rows
ng.head(5)
```

```python
#lists all columns in our data
ng.columns
```

```python
#cleaning data
ng.dropna()
```

```python
#obtaining a list of foods with prices greater 
#than 1600 Naira between May amd December 2019

ng[(ng["May-19"]>1600) & (ng["Dec-19"]>1600)]
```

```python
#splitting highest and lowest columns using delimiters
ng[["Highest","higherprice"]] = ng["Highest"].str.split("(",expand=True)
ng[["higherprice","Blank"]] = ng["higherprice"].str.split(")",expand=True)
ng[["Lowest","lowerprice"]] = ng["Lowest"].str.split("(",expand=True)
ng[["lowerprice","Blank2"]] = ng["lowerprice"].str.split(")",expand=True)
ng.head(5)
```

```python
#deleting blank columns and saving our modified data
del ng["Blank"]
del ng["Blank2"]
ng
ng=ng.to_csv("foodprices.csv",index=False)
```

```python
fp=pd.read_csv("foodprices.csv")
del fp["ItemLabels"]
fp.head(5)
```

```python
# the average of the higher price  and lower price spent on the purchase of food items
fp["higherprice"].mean(), fp["lowerprice"].mean()
```

```python
# getting top 5 states that have high food prices and the frequency.
fp["Highest"].value_counts().head(5)

# Bayelsa is the state with the highest food prices since it has the highest frequency
```

```python
# a plot to see the average price per food item and the spread
# first get the average per food item across the months.
fp["meanperproduct"]=fp.iloc[:,2:42].mean(axis=1)
fp.loc[43] = fp.mean()
fp["RenamedItemLabels"] = fp["RenamedItemLabels"].fillna("monthlymean")
fp.tail(5)
fp.to_csv("foodprices.csv",index=False)
fp.plot(kind="bar",x="RenamedItemLabels",y="meanperproduct")
fp.plot(kind="scatter",x="RenamedItemLabels",y="meanperproduct")
plt.show()
```
