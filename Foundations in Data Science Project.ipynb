# Case Four- Selected food Prices Watch in Nigeria

### Analysis on different food prices and comparison of different food items.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot
```

```python
ng=pd.read_csv("cfafrica-_-data-team-_-wanadata-_-nigeria-_-food-prices-_-april-2020-national.csv")

#reading top 5 rows
ng.head(5)
```

```python
#cleaning data
ng.dropna()
#lists all columns in our data
ng.columns
```

```python
#obtaining a list of foods with prices greater 
#than 1600 Naira between May and December 2019

ng[(ng["May-19"]>1600) & (ng["Dec-19"]>1600)]
```

```python
#splitting highest and lowest columns using delimiters
ng[["Highest","higherprice"]] = ng["Highest"].str.split("(",expand=True)
ng[["higherprice","Blank"]] = ng["higherprice"].str.split(")",expand=True)
ng[["Lowest","lowerprice"]] = ng["Lowest"].str.split("(",expand=True)
ng[["lowerprice","Blank2"]] = ng["lowerprice"].str.split(")",expand=True)
ng.head(5)
```

```python
#deleting blank columns and saving our modified data
del ng["Blank"]
del ng["Blank2"]
ng
ng=ng.to_csv("foodprices.csv",index=False)
```

```python
fp=pd.read_csv("foodprices.csv")
del fp["ItemLabels"]
fp.head(5)
```

```python
# the highest amount  and lowest amount spent on the purchase of food items
fp["higherprice"].max(), fp["lowerprice"].min()
#the highest amont spent on the purchase of  food item 3057.14 and the lowest amount spent on the purchase of  food item is 35.15
```

```python
# getting top 5 states that have high food prices and the frequency.
fp["Highest"].value_counts().head(2), fp["Lowest"].value_counts().tail(2)

# Bayelsa is the most expensive state and Jigawa and Kebbi are amongst the least expensive states
```

```python
# a plot to see the average price per food item and the spread

# first get the average per food item across the months.
fp["meanperproduct"]=fp.iloc[:,2:42].mean(axis=1)
fp.loc[43] = fp.mean()
fp["RenamedItemLabels"] = fp["RenamedItemLabels"].fillna("monthlymean")
fp.tail(5)
fp.to_csv("foodprices.csv",index=False)

#select the products we want to plot
fp.loc[:, "RenamedItemLabels"]
rows=[0,5,9,12,16,22,25,29,35,42]
fpl=fp.loc[rows]
fpl.to_csv("plottingdata.csv",index=False)

#plot
#fpl.plot(kind="bar",x="RenamedItemLabels",y="meanperproduct")
#plt.xlabel("Food Item")
#plt.ylabel("Average Price")
#plt.title("Average Food Prices")
fpl.plot(kind="scatter",x="RenamedItemLabels",y="meanperproduct")
plt.xlabel("Food Item")
plt.ylabel("Average Price")
plt.title("Average Food Prices")
plt.show()
```

```python
#trends over time
#fpt = fpl.T
fpt = fpl.set_index("RenamedItemLabels").T #on csv the title doesn't appear, follow up on this
fpt.to_csv("foodprices_transposed.csv")

series =pd.read_csv('foodprices_grouped.csv', index_col=0)
fig, axes = plt.subplots(nrows=2, ncols=2,constrained_layout=True,figsize=(6, 6))
series['S.Bread'].plot(ax=axes[0,0]); axes[0,0].set_title('S.Bread')
series['D.Catfish'].plot(ax=axes[0,1]); axes[0,1].set_title('D.Catfish')
series['V.Oil'].plot(ax=axes[1,0]); axes[1,0].set_title('V.Oil')
series['R.Plantain'].plot(ax=axes[1,1]); axes[1,1].set_title('R.Plantain')

```
